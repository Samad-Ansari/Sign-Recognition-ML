```python
# SETUP THE VIDEO CAMERA
cap = cv2.VideoCapture(0)
cap.set(3, frameWidth)
cap.set(4, frameHeight)
cap.set(10, brightness)
starting_time = time.time()
frame_id = 0

while(True):
    success, imgOrignal = cap.read()
    frame_id = frame_id + 1
    gray=cv2.cvtColor(imgOrignal,cv2.COLOR_BGR2GRAY)
    co_img=triangle.detectMultiScale(gray,1.3,5)  
    img = np.asarray(imgOrignal)
    img = cv2.resize(img, (32, 32))
    img = preprocessing(img)
    cv2.imshow("Processed Image", img)
    img = img.reshape(1, 32, 32, 1)
    for (x,y,w,h) in co_img:
        image=gray[y:y+w,x:x+w]
        cv2.rectangle(imgOrignal,(x,y),(x+w,y+h),(0, 0, 255),2)
    predictions = model.predict(img)
    classIndex = model.predict_classes(img)
    probabilityValue =np.amax(predictions)
    cv2.putText(imgOrignal, "For better accuracy please keep sign at distance ", (20, 420), font, 0.75, (255, 0, 71), 2, cv2.LINE_AA)
    cv2.putText(imgOrignal, "of 10-20 cm from webcam", (20, 455), font, 0.75, (255, 0, 71), 2, cv2.LINE_AA)
    if probabilityValue > threshold:
            #print(getCalssName(classIndex))
        cv2.putText(imgOrignal, "CLASS :  " , (20, 35), font, 0.75, (0, 0, 255), 2, cv2.LINE_AA)
        cv2.putText(imgOrignal, "PROBABILITY:  ", (20, 75), font, 0.75, (0, 0, 255), 2, cv2.LINE_AA)
        cv2.putText(imgOrignal,str(classIndex)+" "+str(getCalssName(classIndex)), (120, 35), font, 0.75, (255, 0, 92), 2, cv2.LINE_AA)
        cv2.putText(imgOrignal, str(round(probabilityValue*100,2) )+"%", (180, 75), font, 0.75, (255, 0, 92), 2, cv2.LINE_AA)
    else:
        cv2.rectangle(imgOrignal,(10,10),(210,50),(0, 0, 255),2)
        cv2.putText(imgOrignal, "Show the sign", (20, 35), font, 0.75, (255, 255, 255), 2, cv2.LINE_AA)
    
    elapsed_time = time.time() - starting_time
    fps = frame_id / elapsed_time
    cv2.rectangle(imgOrignal,(20,350),(190,390),(0, 0, 255),2)
    cv2.putText(imgOrignal, "FPS : " + str(round(fps, 3)), (30, 380), font, 0.75, (255, 255, 255), 2, cv2.LINE_AA)
    cv2.imshow("Result", imgOrignal)
     
    key=cv2.waitKey(1)
    
    if(key==27):
        break
        
cv2.destroyAllWindows()
cap.release()
```
